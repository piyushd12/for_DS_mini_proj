```python
print(df.dtypes)

numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'FraudFlag' not in numerical_cols:
    numerical_cols.append('FraudFlag')

numerical_df = df[numerical_cols]

correlation_matrix = numerical_df.corr()

fraud_flag_correlation = correlation_matrix['FraudFlag']

threshold = 0.05

low_correlation_features = fraud_flag_correlation[fraud_flag_correlation.abs() < threshold].index.tolist()

print(low_correlation_features)
```

#Findings



1.   FraudFlag is imbalenced wiith with 90.35% of transactions labeled as non-fraudulent and 9.65% as fraudulent.
2.   several features with low correlation to FraudFlag (below a threshold of 0.05): TransactionID, Amount, Latitude, Longitude, AvgTransactionAmount, FailedAttempts, and PhoneNumber.
3. (For this dataset) we need to oversample the minority class (fraudulent transactions) and undersample the majority class (Non-fraudulent transactions)



```python
low_correlation_features.append('DeviceID')
new_df = df.drop(columns=low_correlation_features)

X = new_df.drop('FraudFlag', axis=1)
y = new_df['FraudFlag']
```

```python
from imblearn.over_sampling import SMOTE, SMOTENC
from sklearn.preprocessing import LabelEncoder

categorical_features = X.select_dtypes(include=['object', 'bool']).columns
numerical_features = X.select_dtypes(include=['number']).columns

for feature in categorical_features:
    le = LabelEncoder()
    X[feature] = le.fit_transform(X[feature])

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
resampled_df = pd.DataFrame(X_resampled, columns=X.columns)

resampled_df['FraudFlag'] = y_resampled

print(resampled_df['FraudFlag'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='FraudFlag', data=resampled_df)
plt.title("Class Distribution After SMOTE")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(y, bins=20, kde=True, ax=ax[0])
ax[0].set_title("Before SMOTE")

sns.histplot(y_resampled, bins=20, kde=True, ax=ax[1])
ax[1].set_title("After SMOTE")

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model = XGBClassifier(eval_metric='logloss')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"XGBoost Accuracy: {accuracy:.4f}")
print(f"XGBoost Precision: {precision:.4f}")
print(f"XGBoost Recall: {recall:.4f}")
print(f"XGBoost F1-score: {f1:.4f}")
```
